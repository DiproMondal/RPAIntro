```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options


import pandas as pd
```

```python
def item_data(query = 'Nike', pages = 2):
    options = Options()
    options.headless=True
    url = 'https://www.amazon.com/'
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    driver.find_element_by_xpath("//input[@id='twotabsearchtextbox']").send_keys(query)
    driver.find_element_by_xpath("//input[@type='submit']").click()
    WebDriverWait(driver, 10) ## This will make the script wait for 10 ms.
    driver.find_element_by_xpath("//*[@id='p_72/2661618011']/span/a/section/span").click()
    WebDriverWait(driver, 10000).until(lambda x: x.find_element_by_xpath("//*[contains(text(), 'Next')]"))

    Href = []
    Items = []
    Curr = []
    Pric = []
    GlbR = []
    AvgR = []
    for p in range(0, pages):
        all_items = driver.find_elements_by_class_name('s-result-item')
        hrefs = [''] * len(all_items)
        item_name = [''] * len(all_items)
        currency = [''] * len(all_items)
        price = [''] * len(all_items)
        gl_rt = [''] * len(all_items)
        ag_rt = [''] * len(all_items)
        for i in range(0, len(all_items)):
            try:
                hrefs[i] =all_items[i].find_element_by_tag_name('a').get_attribute('href')
            except Exception:
                pass
            try:
                item_name[i] = all_items[i].find_element_by_class_name('a-text-normal').text
            except Exception:
                pass
            try:
                currency[i] = all.items[i].find_element_by_class_name('a-price-symbol').text 
            except Exception:
                pass
            try:
                price[i] = all_items[i].find_element_by_class_name('a-price-whole').text + '.' + all_items[5].find_element_by_class_name('a-price-fraction').text 
            except Exception:
                pass
            try:    
                gl_rt[i] = all_items[i].find_element_by_xpath("//*[@class = 'a-size-base']").text
            except Exception:
                pass
            try:    
                ag_rt[i] = all_items[i].find_element_by_class_name('a-icon-alt').get_attribute('textContent')
            except Exception:
                pass
            
        Href.append(hrefs)
        Items.append(item_name)
        Curr.append(currency)
        Pric.append(price)
        GlbR.append(gl_rt)
        AvgR.append(ag_rt)
        
        # Need to wait for the Next to be visible otherwise it will return error
        WebDriverWait(driver, 10000).until(lambda x: x.find_element_by_xpath("//*[contains(text(), 'Next')]"))
        driver.find_element_by_xpath("//*[contains(text(), 'Next')]").click()
        WebDriverWait(driver, 10000).until(lambda x: x.find_element_by_xpath("//*[contains(text(), 'Next')]"))
        
    Href = [item for l in Href for item in l]
    Items = [item for l in Items for item in l]
    Curr = [item for l in Curr for item in l]
    Pric = [item for l in Pric for item in l]
    GlbR = [item for l in GlbR for item in l]
    AvgR = [item for l in AvgR for item in l]
    
    df = pd.DataFrame(list(zip(Items,Href, Curr, Pric, GlbR, AvgR)), columns=['Item_Name','Link', 'Currency', 'Price', 'Global_Ratings', 'Average_Ratings'])
    return(df)
```

```python
data = item_data()
```
